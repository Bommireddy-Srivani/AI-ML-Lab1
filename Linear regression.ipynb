```python
import pandas as pd
```

```python
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
data=pd.read_csv("C:\\Users\\SRIVANI\\Downloads\\archive (5)\\SOCR-HeightWeight.csv")
data.head()
```

```python
data.info()
```

```python
data.nunique()
```

```python
data.plot(x="Height(Inches)",y="Weight(Pounds)",kind="scatter",color="blue")
plt.title("Relation between height and weight")
plt.xlabel("Height")
plt.ylabel("Weight")
plt.show()
```

```python
x = data[["Height(Inches)"]].values  
y = data["Weight(Pounds)"].values    
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
print(x_train.shape)
print(y_train.shape)
```

```python
model=LinearRegression()
```

```python
model.fit(x_train,y_train)
y_pred=model.predict(x_test)
```

```python
mae=mean_absolute_error(y_test,y_pred)
print(mae)
```

```python
# Plot predictions
plt.scatter(x_test, y_test, color="blue", label="Actual")
plt.plot(x_test, y_pred, color="red", linewidth=2, label="Predicted")
plt.title("Height vs Weight Prediction")
plt.xlabel("Height (Inches)")
plt.ylabel("Weight (Pounds)")
plt.legend()
plt.show()
```
