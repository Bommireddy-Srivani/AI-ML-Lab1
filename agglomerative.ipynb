```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.cluster.hierarchy as sch
```

```python
data_points=np.array([[0.40,0.53],[0.22,0.38],[0.35,0.32],[0.26,0.19],[0.08,0.41],[0.45,0.30]])
labels=['p1','p2','p3','p4','p5','p6']
```

```python

linkages=['single','complete','average','weighted','centroid','median','ward']
```

```python
for link in linkages:
    linked=sch.linkage(data_points,method=link,metric='euclidean')
    plt.figure(figsize=(10,5))
    sch.dendrogram(linked,labels=labels,orientation='top',distance_sort='ascending',show_leaf_counts=True)
    plt.title(f"Agglomerative Clustering Dendrgram ({link} Linkage)")
    plt.xlabel("Sample points")
    plt.ylabel("Distane")
    plt.grid(False)
    plt.tight_layout()
    plt.show()
```
